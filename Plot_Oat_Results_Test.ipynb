```python
Oat_Data = pd.read_csv("Oat_Data.csv")
```

```python
Oat_Data.Date=pd.to_datetime(Oat_Data.Date, format="%m/%d/%Y")
```

```python
Days = Oat_Data.Date -Oat_Data.Date.min()
```

```python
Days
```

```python
Oat_Data["Days"] = Days.dt.days ##Days as integer
```

```python
sns.set_theme(style ="whitegrid")

ax1 = sns.lineplot(data = Oat_Data, x="Days", y="Vegetation_Fraction",   
                   hue="Cover", palette = ["r", "b", "g"], style = "Cover", markers = True)
#ax1.set(xticks=ticks)
#ax1.set_xticklabels(["7/8", "8/21", "9/9", "9/29", "10/5", "10/15", "10/25"])
#ax1.set_ylabels("Vegetation Cover")
ax1.set(xlabel = "Date", ylabel = "Vegetation Cover",  
        ylim=(0,1.1), xlim=(-10, 120), 
        yticklabels = (["0%", "20%", "40%", "60%", "80%", "100%"]))
ax1.set_xticklabels(labels = ["6/18", "7/8", "7/28", "8/17", "9/6", "9/26", "10/16", "11/5"], rotation= 90)
ax1.axvline(29, color='dimgrey', ls = "--", label = "Harvest")
ax1.axvline(35, color='dimgrey', ls = ":", label = "Post Seed")
ax1.set_title("Percent Vegetation", size = 20)
ax1.legend(bbox_to_anchor=(1,1))

#ax1.axes.flat[0].yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1))


```
